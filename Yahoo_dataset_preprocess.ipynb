```python
import glob
import datetime
import pandas as pd
import numpy as np
from math import sqrt
from sklearn.preprocessing import MinMaxScaler,StandardScaler

a1_csv = glob.glob(f'./data/ydata-labeled-time-series-anomalies-v1_0/A4Benchmark/*.csv', recursive=True)
print(f'Loaded the paths of {len(a1_csv)} files from disk. Begin processing at: {datetime.datetime.now()}')

for index,file in enumerate(a1_csv):
    if index%10 == 0:
        print(f'Processing index: {index} of {len(a1_csv)}')
        
    fname = file.split("/")[4].replace('\\','').split(".")[0]
    print(fname)
    df = pd.read_csv(file)
#     df['timestamp'] = pd.to_datetime(df['timestamp'],unit='s')
#     df_indexed = df.set_index('timestamp')

#     print("Data:")
#     print(df_indexed.head())

    scaler2 = StandardScaler()
    stddf = df.copy()
    for col in stddf.columns:
        if col not in ["timestamp","is_anomaly"]:
            stddf[col] = scaler2.fit_transform(np.reshape(df[col].values,(len(df[col].values),1)))

#     print('Mean: %f, StandardDeviation: %f' % (scaler2.mean_, sqrt(scaler2.var_)))
#     print("Standardized values:")
#     print(stddf.head()) 

    scaler = MinMaxScaler(feature_range=(0, 1))
    #scaler = scaler.fit(np.reshape(masked[col].values,(len(masked[col]),1)))        
    scaleddf = stddf.copy()
    for col in scaleddf.columns:
        if col not in ["timestamp","is_anomaly"]:
            scaleddf[col] = scaler.fit_transform(np.reshape(df[col].values,(len(df[col].values),1)))
#     print("Normalized values:")
#     print(scaleddf.head())
    
    fileName = "./data/ydata-labeled-time-series-anomalies-v1_0/A4Benchmark_processed/" + fname + ".csv"
    scaleddf.to_csv(fileName, index=False)
    
#     print(f'iteration for index {index} finished\n')
    
    

#     print("\n\n\n")
```
